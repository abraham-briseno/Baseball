```python
# Import libraries
import pandas as pd
import seaborn as sns
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_sample_weight
```

# Read and Clean data


```python
pitches = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pitches.csv")
pd.set_option('display.max_columns', None)
pitches
```

```python
# Chronological daat to predict pitch
pitches = pitches.sort_values(by=['game_pk', 'inning', 'top', 'at_bat_num', 'pcount_at_bat'])

# Create the target variables and previous pitch
pitches['next_pitch_type'] = pitches.groupby(['game_pk', 'at_bat_num'])['pitch_type'].shift(-1)
pitches['previous_pitch_type'] = pitches.groupby(['game_pk', 'at_bat_num'])['pitch_type'].shift(1)

# Drop columns and data that may leak or is irrelevant
columns_to_drop = [
    'start_speed', 'end_speed', 'sz_top', 'sz_bot', 'pfx_x', 'pfx_z',
    'px', 'pz', 'x0', 'z0', 'y0', 'vx0', 'vz0', 'vy0', 'ax', 'az', 'ay',
    'break_length', 'break_y', 'break_angle', 'spin_dir', 'spin_rate',
    'type_confidence', 'zone', 'nasty',

    'type', 'event', 'event2', 'event3', 'event4', 'pitch_des',
    'at_bat_des', 'is_final_pitch', 'final_balls', 'final_strikes', 'final_outs',

    'uid', 'game_pk', 'year', 'team_id_b', 'team_id_p',
    'start_tfs', 'start_tfs_zulu', 'pitch_tfs', 'pitch_tfs_zulu',
    'sv_id', 'pitch_id', 'x', 'y', 'created_at', 'added_at',
    'modified_at', 'modified_by', 'at_bat_num'
]


# any errors in the list will be ignored
pitches_cleaned = pitches.drop(columns=columns_to_drop, errors='ignore')

runner_cols = [col for col in pitches_cleaned.columns if col.startswith('runner')]
pitches_cleaned = pitches_cleaned.drop(columns=runner_cols, errors='ignore')

# Drop original 'pitch_type' to prevent leakeage essentailly for the model not to teach
pitches_cleaned = pitches_cleaned.drop(columns=['pitch_type'], errors='ignore')

# Drop NaN/Na beofre training the model
pitches_cleaned = pitches_cleaned.dropna(subset=['next_pitch_type'])
print(pitches_cleaned.columns.tolist())
```

```python
# Name the target groups(will help later to interpret results)
pitch_category_map = {
    # Fastballs (Hard)
    'FF': 'Fastball', 'FT': 'Fastball', 'SI': 'Fastball', 'FC': 'Fastball',
    # Breaking Balls (Movement)
    'SL': 'Breaking', 'CU': 'Breaking', 'KC': 'Breaking',
    # Off-speed (Deception)
    'CH': 'Offspeed', 'FS': 'Offspeed', 'KN': 'Offspeed', 'EP': 'Offspeed'
}

# Create new columns for the groups
pitches_cleaned['next_pitch_group'] = pitches_cleaned['next_pitch_type'].map(pitch_category_map)
pitches_cleaned['previous_pitch_group'] = pitches_cleaned['previous_pitch_type'].map(pitch_category_map)
pitches_cleaned = pitches_cleaned.dropna(subset=['next_pitch_group'])
```

```python
pitches_cleaned
```

# Prepare and Encode data

```python
pitches_cleaned['previous_pitch_type'] = pitches_cleaned['previous_pitch_type'].fillna('First_Pitch')

# Convert base runner to binary values
for col in ['on_1b', 'on_2b', 'on_3b']:
    pitches_cleaned[col] = pitches_cleaned[col].notna().astype(int)

# Convert height to integers
def parse_height(height_str):
    if pd.isna(height_str):
        return 0
    parts = str(height_str).split('-')
    return int(parts[0]) * 12 + int(parts[1]) if len(parts) == 2 else 0

pitches_cleaned['b_height'] = pitches_cleaned['b_height'].apply(parse_height)
```

```python
cols_to_drop_now = ['date', 'batter_id', 'pitcher_id']
pitches_cleaned = pitches_cleaned.drop(columns=cols_to_drop_now, errors='ignore')
```

```python
categorical_cols = pitches_cleaned.select_dtypes(include=['object', 'category', 'string']).columns.tolist()

if 'next_pitch_type' in categorical_cols:
    categorical_cols.remove('next_pitch_type')
if 'next_pitch_group' in categorical_cols:
    categorical_cols.remove('next_pitch_group')

# Filter columns by 20 unique values
low_cardinality_cols = [col for col in categorical_cols if pitches_cleaned[col].nunique() < 20]
high_cardinality_cols = list(set(categorical_cols) - set(low_cardinality_cols))

print(f"Good: Safe to one-hot encode: {low_cardinality_cols}")
print(f"Bad: Dropping due to high cardinality (too many unique values): {high_cardinality_cols}")

# Get dummy variables
pitches_model = pd.get_dummies(pitches_cleaned, columns=low_cardinality_cols, drop_first=True)
pitches_model = pitches_model.drop(columns=high_cardinality_cols, errors='ignore')

# 5. Convert booleans to floats
pitches_model = pitches_model.astype(float, errors='ignore')


print("\nEncoding complete! New dataset shape:", pitches_model.shape)
```

```python
import gc
# Extract variables we want
y_granular = pitches_model['next_pitch_type']
y_grouped = pitches_model['next_pitch_group']

# Drop targets from the dataset
X = pitches_model.drop(columns=['next_pitch_type', 'next_pitch_group'])

del pitches_model
gc.collect()

# Split the data
# Using random_state=42 ensures X_train perfectly aligns with both y_train targets
X_train, X_test, y_train_gran, y_test_gran = train_test_split(X, y_granular, test_size=0.2, random_state=42)
_, _, y_train_grp, y_test_grp = train_test_split(X, y_grouped, test_size=0.2, random_state=42)
```

# Random Forest Model

```python
# Model 1 with 17 pitch types
rf_granular = RandomForestClassifier(n_estimators=25, max_depth=15, class_weight='balanced', random_state=42, n_jobs=-1)
rf_granular.fit(X_train, y_train_gran)

preds_gran = rf_granular.predict(X_test)
print("\nGRANULAR MODEL RESULTS")
print("Accuracy:", accuracy_score(y_test_gran, preds_gran))
print(classification_report(y_test_gran, preds_gran))

# Model 2 The Grouped Model with 3 Categories
print("\nGrouped Model (3 Categories)...")
rf_grouped = RandomForestClassifier(n_estimators=25, max_depth=15, class_weight='balanced', random_state=42, n_jobs=-1)
rf_grouped.fit(X_train, y_train_grp)

preds_grp = rf_grouped.predict(X_test)
print("\nGROUPED MODEL RESULTS")
print("Accuracy:", accuracy_score(y_test_grp, preds_grp))
print(classification_report(y_test_grp, preds_grp))
```

Granular and Grouped Models are the two models tha are the
classification reports that evaluates the model's predictive performance. The dataset contains a total of 104,920 pitches whhich is the support. The model above predicts the specific pitch types e.g., Four-Seam Fastball, Slider, Curveball.

To improve the models performance the pitch types were grouped  into three broader categories Breaking, Fastball, and Offspeed.

Category Breakdown:
Fastball: Represents the majority of pitches at 64,281 and in baseball this is expe ted. The model has strong precision here of 0.71 but struggles with recall at 0.36 meaning when it guesses a Fastball it is usually is right but it misses a large portion of actual fastballs.

Breaking: Contains 26,449 pitches in this category it is moderately balanced with a 0.35 precision and 0.50 recall.

```python
# Find the probabilities
transition_matrix = pd.crosstab(
    index=pitches_cleaned['previous_pitch_group'],
    columns=pitches_cleaned['next_pitch_group'],
    normalize='index'
)
transition_matrix = (transition_matrix * 100).round(1)
print("Pitch Sequencing Probabilities (Row = Previous Pitch, Column = Next Pitch):")
print(transition_matrix)

transition_matrix.to_csv('Pitch_Transition_Matrix.csv')
```

Pitch Sequencing Probabilities

The probabilites displays a transition matrix showing the likelihood of the next pitch category based on the previous pitch category.

The Fastball First Rule: Regardless of what the previous pitch was a fastball is consistently the most probable next pitch. If the previous pitch was a Breaking ball there is a 56.0% chance the next pitch is a Fastball. If it was Offspeed there is a 55.9% chance of a Fastball to be next.

Pitch Repetition: Pitchers are highly likely to double down on fastballs. If a pitcher throws a fastball there is a 63.6% probability that the next pitch will also be a fastball. Breaking balls are repeated 34.3% of the time and Offspeed pitches are repeated 25.0% of the time.

Rarest Sequences: An Offspeed pitch following a Breaking pitch is the least likely sequence occurring only 9.7% of the time. Similarly an Offspeed pitch following a Fastball only happens 12% of the time.

Final interpretation: This is an excellent way to decide quickly what the probability of the next pitch will be. Although it doesn not predict the specific pitch it predicts a category but it tremendously helps batters to realize if they need to adjust to a quicker swing or a slower swing with better timing and swing accuracy.

# XGBoost Model

```python
# Granular Setup for 17 Pitches
le_gran = LabelEncoder()
y_train_gran_num = le_gran.fit_transform(y_train_gran)
y_test_gran_num = le_gran.transform(y_test_gran)
weights_gran = compute_sample_weight(class_weight='balanced', y=y_train_gran_num)

# Grouped Setup 3 Categories
le_grp = LabelEncoder()
y_train_grp_num = le_grp.fit_transform(y_train_grp)
y_test_grp_num = le_grp.transform(y_test_grp)
weights_grp = compute_sample_weight(class_weight='balanced', y=y_train_grp_num)

# TRAIN MODEL for 17 PITCHES
xgb_gran = xgb.XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.1, random_state=42, n_jobs=-1)
xgb_gran.fit(X_train, y_train_gran_num, sample_weight=weights_gran)

# Predict and decode back to text
preds_gran_num = xgb_gran.predict(X_test)
preds_gran_text = le_gran.inverse_transform(preds_gran_num)

print("\nXGBOOST RESULTS 17 Pitches")
print("Accuracy:", accuracy_score(y_test_gran, preds_gran_text))
print(classification_report(y_test_gran, preds_gran_text))

# TRAIN MODEL for 3 CATEGORIES
xgb_grp = xgb.XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.1, random_state=42, n_jobs=-1)
xgb_grp.fit(X_train, y_train_grp_num, sample_weight=weights_grp)

# Predict and decode back to text
preds_grp_num = xgb_grp.predict(X_test)
preds_grp_text = le_grp.inverse_transform(preds_grp_num)

print("\nXGBOOST GROUPED RESULTS")
print("Accuracy:", accuracy_score(y_test_grp, preds_grp_text))
print(classification_report(y_test_grp, preds_grp_text))


# Features for grouped XGBoost model
xgb_importances = xgb_grp.feature_importances_

xgb_feature_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance_Score': xgb_importances
}).sort_values(by='Importance_Score', ascending=False)

print("\nTop 15 XGBoost Pitch Prediction Drivers (Grouped Model):")
print(xgb_feature_df.head(15))

xgb_feature_df.to_csv('XGBoost_Pitch_Drivers.csv', index=False)
```

17 Pitches

The dataset evaluates a broad spectrum of pitch types and they are categorized under the 17 Pitches model. Predicting exact pitch variations from such diverse options is incredibly difficult because pitchers actively employ deception, tunneling, and randomized sequencing to disrupt a hitter's timing. While traditional machine learning metrics might view the baseline accuracy as low, the model is successfully learning the underlying signals and pitcher tendencies rather than just randomly guessing across imbalanced data.

Grouped Results
When simplifying the classification into three broader categories (Fastball, Breaking, and Offspeed), the overall model accuracy increases to 38.5%. This grouped approach is highly practical because it mirrors how a hitter processes information in the batter's box. They often look for a general speed and movement of the ball or movement of the pitchers arm/hand rather than needing to identify the exact pitch type.
Fastball Precision: Fastballs comprise the majority of the dataset (64,281 pitches) and attain a highly actionable precision of 0.74. This is the most valuable metric because it means when the model predicts a fastball it is correct 74% of the time allowing a hitter to confidently expect a high speed ball.

Offspeed Recall: Offspeed pitches demonstrate the highest recall in this group at 0.63 though they have the lowest precision at 0.21.


Prediction Drivers
The mathematical features driving the XGBoost algorithm reveal that the model has effectively learned baseball psychology and game theory validating its situational awareness much like any player.

Sequencing Tactics: The most heavily weighted feature for the grouped model is "previous_pitch_group_Offspeed" at (0.228225). This perfectly illustrates the strategic pitching loop where an offspeed pitch is often followed by a fastball to exploit a batter's slowed down bat speed.

Physiological Matchups: The pitcher's handedness ("p_throws_R") is the second most important driver at (0.166765). This highlights the importance of advantages as the physical orientation of the pitcher dictates the sequence and arsenal used against specific batters.

Game State Context: The model intelligently factors in the macro and micro game states by utilizing immediate count variables like "strikes" and "balls," as well as the current "inning."

```python
# Encode targets into integers
label_encoder_gran = LabelEncoder()
y_train_num_gran = label_encoder_gran.fit_transform(y_train_gran)
y_test_num_gran = label_encoder_gran.transform(y_test_gran)
sample_weights_gran = compute_sample_weight(class_weight='balanced', y=y_train_num_gran)

# XGBoost Classifier
xgb_model_gran = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    random_state=42,
    n_jobs=-1
)

# Train model
xgb_model_gran.fit(X_train, y_train_num_gran, sample_weight=sample_weights_gran)

# Make predictions
preds_num_gran = xgb_model_gran.predict(X_test)
preds_text_gran = label_encoder_gran.inverse_transform(preds_num_gran)

# Evaluate the results
print("\nXGBOOST MODEL RESULTS")
print("Accuracy:", accuracy_score(y_test_gran, preds_text_gran))
print(classification_report(y_test_gran, preds_text_gran))


# Features importances from XGBoost
xgb_importances_gran = xgb_model_gran.feature_importances_

xgb_feature_df_gran = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance_Score': xgb_importances_gran
}).sort_values(by='Importance_Score', ascending=False)

print("\nTop 15 XGBoost Granular Pitch Prediction Drivers:")
display(xgb_feature_df_gran.head(15))

xgb_feature_df_gran.to_csv('XGBoost_Granular_Pitch_Drivers.csv', index=False)
```

**Final Decision**

The Random Forest model is the better choice for predicting the next pitch.

Superior Accuracy: A 41.6% accuracy in a 3-class baseball prediction model is notably strong given the randomness and many variables of the sport. It comfortably beats XGBoost's 38.5%.

Robust Feature Selection: Random Forest's reliance on a blend of fatigue (pcount), game context (inning), and matchups (stand_R, b_height) creates a more resilient model. XGBoost's heavy reliance on the immediate previous pitch makes it highly susceptible to being tricked or fooled if a pitcher breaks their typical sequencing pattern.

Actionable Balance: While XGBoost's 74% fastball precision is tempting, sacrificing that much recall limits its utility. Random Forest's 71% precision is still highly actionable for a hitter looking to expect on a something fast and it successfully identifies a much larger volume of those opportunities.
